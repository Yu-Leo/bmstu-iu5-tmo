## Загрузка и первичный анализ

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import ast

df = pd.read_csv("/dataset.csv")
df.info()
```

```python
df.head()
```

## Очистка

```python
df.drop(['Employer', 'Name', 'Description', 'Professional roles', 'Profarea names', 'Area', 'Specializations', 'Published at'], axis=1, inplace=True)
```

```python
df.info()
df.head()
```

## Анализ наличия ЗП в вакансии

```python
value_counts = df['Salary'].value_counts()

plt.figure(figsize=(7, 7))
value_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title('Наличие ЗП в описании вакансии')

plt.ylabel('')
plt.show()
```

## Определение наиболее популярного вида занятости

```python
schedule_value_counts = df['Schedule'].value_counts()

plt.figure(figsize=(7, 7))
schedule_value_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, labels=None)

plt.title('Виды занятости в вакансиях')

plt.ylabel('')
plt.legend(labels = ['Полный день', 'Удалённая работа', 'Гибкий график', 'Смешанный график', 'Вахтовый метод'], loc='lower right')

plt.show()
```

## Определение наиболее частно встречающегося в вакансиях навыка

```python
all_keys = []
for i in df['Keys']:
    a = ast.literal_eval(i)
    all_keys.extend(a)
```

```python
keys_counts = Counter(all_keys)

top_10_keys = keys_counts.most_common(10)

keys, counts = zip(*top_10_keys)

plt.figure(figsize=(10, 6))
plt.bar(keys, counts)

plt.title('Наиболее часто встречающиеся навыки')
plt.xlabel('Навык')
plt.ylabel('Частота')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

## Определение корреляции между максимальной заработной платой и требуемым опытом.

```python
experience_order = ['Нет опыта', 'От 1 года до 3 лет', 'От 3 до 6 лет', 'Более 6 лет']
df['Experience'] = pd.Categorical(df['Experience'], categories=experience_order, ordered=True)

plt.figure(figsize=(8, 6))
df.boxplot(column='To', by='Experience', grid=False)

plt.title('ЗП для различного уровня опыта')
plt.suptitle('')
plt.xlabel('Уровень опыта')
plt.ylabel('Максимальная ЗП (руб)')

plt.tight_layout()
plt.show()
```

## Исследовании зависимости между количеством навыков в описании вакансии и максимальной ЗП

```python
df['keys_count'] = df['Keys'].apply(lambda x: len(ast.literal_eval(x)))
```

```python
# Строим scatter plot
plt.figure(figsize=(8, 6))
plt.scatter( df['keys_count'], df['To'], color='blue', edgecolor='black', alpha=0.7)

# Добавляем заголовок и подписи
plt.title('Зависимость между максимальной ЗП и кол-вом навыков')
plt.ylabel('ЗП')
plt.xlabel('Количество навыков')

# Показываем график
plt.tight_layout()
plt.show()
```
